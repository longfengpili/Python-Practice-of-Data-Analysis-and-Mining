```python
result_all = data[(data.TARGET_ID == 184)]
```

```python
def attr_trans(x):
    result_trans = Series(index=['SYS_NAME','CWXT_DB:184:C:\\','CWXT_DB:184:D:\\','COLLECTTIME'])
    result_trans['SYS_NAME'] = x['SYS_NAME'].iloc[0]
    result_trans['COLLECTTIME'] = x['COLLECTTIME'].iloc[0]
    result_trans['CWXT_DB:184:C:\\'] = x['VALUE'].iloc[0]
    result_trans['CWXT_DB:184:D:\\'] = x['VALUE'].iloc[1]
    
    return result_trans

x = result_all.groupby('COLLECTTIME')
```

```python
result_t = result_all.groupby('COLLECTTIME').apply(attr_trans)
result_t[:5]
```

```python
result_t.loc[:,['CWXT_DB:184:C:\\','CWXT_DB:184:D:\\']].plot(kind='line',style='o--',grid=True,figsize=(10,5))
```

```python
result_t_d = result_t.loc[:,'CWXT_DB:184:D:\\'][:-5]
#自相关图/偏相关图
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
plot_acf(result_t_d).show()
#plot_pacf(result_t_d).show()

from statsmodels.tsa.stattools import adfuller as ADF
diff = 0
adf = ADF(result_t_d) #依次返回为adf、pvalue(P值)、usedlag/nobs/critical values/icbest/regresults/resstore
while adf[1] >= 0.05: #p值小于0.05认为是正常
    diff += 1
    adf = ADF(result_t_d.diff(diff).dropna())
    
print('经过{}阶差分后正常,各项数据为{}'.format(diff,adf))

print('--'*10)
plot_acf(result_t_d.diff(diff).dropna()).show()
plot_pacf(result_t_d.diff(diff).dropna()).show()
```

```python
#白噪声检测
from statsmodels.stats.diagnostic import acorr_ljungbox

[lb],[p] = acorr_ljungbox(result_t_d,lags=1)
if p < 0.05:
    print('该原始序列为非白噪音序列,对应p值为{}'.format(p))
else:
    print('该原始序列为白噪音序列,对应p值为{}'.format(p))
   
[lb],[p] = acorr_ljungbox(result_t_d.diff(1).dropna(),lags=1)
if p < 0.05:
    print('一阶差分该序列为非白噪音序列,对应p值为{}'.format(p))
else:
    print('一阶差分该序列为白噪音序列,对应p值为{}'.format(p))
```
