```python
filename = './原书代码数据/chapter9/demo/data/moment.csv'
f = open(filename,'r')
data = pd.read_csv(f,sep=',')
f.close()
data
```

```python
data['类别'].value_counts().sort_index()
data_value = data.values
from numpy.random import shuffle
shuffle(data_value)
data_train = data_value[:int(len(data)*0.8),:]
data_test = data_value[int(len(data)*0.8):,:]
data_train
DataFrame(data_train)[0].value_counts()
```

# 构建支持向量机

```python
#构造特征和标签
x_train = data_train[:,2:]*30 #放大特征
y_train = data_train[:,0].astype(int)

x_test = data_test[:,2:]*30 #放大特征
y_test = data_test[:,0].astype(int)
```

```python
from sklearn import svm
model = svm.SVC()
model.fit(x_train,y_train)

#保存模型
import pickle
pickle.dump(model,open('./原书代码数据/chapter9/demo/tmp/svm_xu.model','wb'))
#可以通过下面的语句load模型
#model = pickle.load(open('./原书代码数据/chapter9/demo/tmp/svm_xu.model','rb'))
```

```python
model.predict(x_train)
```

```python
train_predict = pd.concat([DataFrame(data_train),DataFrame(model.predict(x_train))],axis=1)
train_predict.columns = list(data.columns) + ['predict']
train_predict.head()
```

```python
from sklearn import metrics
cm_train = metrics.confusion_matrix(y_train,model.predict(x_train)) #训练样本的混淆矩阵
```
